```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('data/2019-04-25.csv')
```

```python
mkt_value = data.groupby('team_name').median()['market_value'] \
    .reset_index() \
    .sort_values('market_value', ascending=True)
```

```python
mkt_value.plot(kind='barh', x='team_name', y='market_value', figsize=(8,8))
```

```python
plt.figure(figsize=(5,15))
sns.swarmplot(data=data, x='market_value', y='team_name', dodge=True)
```
